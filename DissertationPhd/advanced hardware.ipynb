#### <span style="color:#838383">Построение ненормального шума из случайного блуждания. Наверное, использую в полном моделировании для описания шума высокочастотного сигнала антенн связи.</span>

```python
import matplotlib.pyplot as plt
import numpy as np

def clip(a: float, bot: float, top: float) -> float:
    if a < bot:
        return bot
    if a > top:
        return top
    return a

def get_walk_noise(n: int, err: float = 1., step: float = 0.03) -> list:
    """Функция возвращает вектор длинны n с компонентами, значения которого в диапазоне [-err, err]
    :param n: Длина возвращаемого вектора
    :param err: Максимальный разброс значений
    :param step: Шаг блуждания шума
    :return: Вектор шума длины n"""
    y = [0.]
    for _ in range(n-1):
        y += [clip(y[-1] + np.random.choice([-step, step]), -err, err)]
    return y

n = 10000
plt.figure(figsize=(15,3))
plt.plot([i for i in range(n)], get_walk_noise(n))
plt.grid()
plt.show()
```
