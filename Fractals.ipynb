```python
#Zn+1=Zn*t*(1-Zn)
import random as r
import matplotlib.pyplot as p
import math as mt
import numpy as np

start = 0.5
R = 5000
t = np.linspace(1, 4, R)
d = np.linspace(0, 1, R)
for i in range(R):
    tmp = start
    for j in range(30+int(r.random()*10)):
        #print(t[i], tmp)
        tmp = tmp*t[i]*(1-tmp);
    d[i] = tmp

p.scatter(t, d, color = 'k', marker = '.')
p.show()
```

```python
N = 500
n = 10
X = np.linspace(-2, 2, N)
Y = np.linspace(-2, 2, N)

def z2(z):
    return np.array([(z[0]**2 - z[1]**2), (2*z[0]*z[1])])

def ro2(z):
    return z[0]**2 + z[1]**2

for i in range(N):
    for j in range(N):
        c = np.array([X[i], Y[j]])
        z = np.array([0,0])
        for m in range(n):
            z = z2(z) + c
            #print('x,y = ',i, j, ' z = ', z)
        if ro2(z)<3:
            p.scatter(X[i], Y[j], color = 'k', marker = '.')
            
p.show()
```
