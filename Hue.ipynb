```python
[Ur,Ir,Vr] = np.linalg.svd(pix[:,:,0])
[Ug,Ig,Vg] = np.linalg.svd(pix[:,:,1])
[Ub,Ib,Vb] = np.linalg.svd(pix[:,:,2])

N = 500

Ur1 = np.zeros((dx, N))
Ir1 = np.zeros((N, N))
Vr1 = np.zeros((N, dy))

Ug1 = np.zeros((dx, N))
Ig1 = np.zeros((N, N))
Vg1 = np.zeros((N, dy))

Ub1 = np.zeros((dx, N))
Ib1 = np.zeros((N, N))
Vb1 = np.zeros((N, dy))

for i in range(dx):
    for j in range(N):
        Ur1[i][j] = Ur[i][j];
        Ug1[i][j] = Ug[i][j];
        Ub1[i][j] = Ub[i][j];
        
for i in range(N):
    Ir1[i][i] = Ir[i];
    Ig1[i][i] = Ig[i];
    Ib1[i][i] = Ib[i];
        
for i in range(N):
    for j in range(dy):
        Vr1[i][j] = Vr[i][j];
        Vg1[i][j] = Vg[i][j];
        Vb1[i][j] = Vb[i][j];

imgr = (Ur1.dot(Ir1)).dot(Vr1);
imgg = (Ug1.dot(Ig1)).dot(Vg1);
imgb = (Ub1.dot(Ib1)).dot(Vb1);
img = np.zeros((dx, dy, 3));
for i in range(dx):
    for j in range(dy):
        img[i][j][0] = imgr[i][j];
        img[i][j][1] = imgg[i][j];
        img[i][j][2] = imgb[i][j];
        
        
img2 = Image.fromarray(img, 'RGB')

plt.imshow(img2)
```
