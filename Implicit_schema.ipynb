```python
import matplotlib.pyplot as p
import math
import numpy as np

# Задание констант
begin = 0
end = math.pi
N = 1000
h = (end - begin)/N

# Задание коэффициентов разностной задачи для метода прогонки
x = np.linspace(begin, end, N)
d = list(range(N))
d[0] = 0
d[N - 1] = math.pi**2
a = list(range(N-1))
a[N-2] = 0
b = list(range(N))
b[0] = 1
b[N - 1] = 1
c = list(range(N-1))
c[0] = 0

# Подсчёт коэффициентов
for i in range(N - 2):
    d[i+1] = (2 + 2*x[i+1]**3 - 6*x[i+1] + (x[i+1]**2 - 3)*math.exp(x[i+1])*math.sin(x[i+1]*(1+math.cos(x[i+1]))) + math.cos(x[i+1])*(math.exp(x[i+1])*(x[i+1]**2 - 1) + x[i+1]**4 - 3*x[i+1]**2))
    b[i+1] = -2 - (x[i+1]**2 - 3)*h - (x[i+1]**2 - 3)*h**2*math.cos(x[i+1])
    a[i+1] = 1
    c[i+1] = 1 + (x[i+1]**2 - 3)*h

# Подсчёт p и q
p = list(range(N))
q = list(range(N))
p[0] = -c[0]/b[0]
q[0] = d[0]/b[0]

for i in range(N - 1):
    p[i+1] = -c[i]/(a[i]*p[i] + b[i])
    q[i+1] = (d[i] - a[i]*q[i])/(a[i]*p[i] + b[i])

# Подсчёт разностной задачи
y = np.linspace(begin, end, N)
y[0] = 0
y[N - 1] = math.pi**2

for i in reversed(range(N - 3)):
    y[N - 2] = y[N - 1]*p[N - 1] + q[N - 1]
    print(i, y[N - 2], x[N - 2])
    
for i in range(N):
    if (abs(x[i] - 0.5) < h):
        print(i, x[i],y[i])
    if (abs(x[i] - 1) < h):
        print(i, x[i],y[i])
    if (abs(x[i] - 1.5) < h):
        print(i, x[i],y[i])
    if (abs(x[i] - 2) < h):
        print(i, x[i],y[i])
    if (abs(x[i] - 2.5) < h):
        print(i, x[i],y[i])
    if (abs(x[i] - 3) < h):
        print(i, x[i],y[i])
```
