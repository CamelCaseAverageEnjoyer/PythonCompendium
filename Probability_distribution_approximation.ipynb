```python
import random as r
import math
import matplotlib.pyplot as p

N = 100
U = 1000000

a = list(range(N))
x = list(range(N))
for i in range(N):
    x[i] = 0
for i in range(U):
    tmp = r.random()*N
    x[math.floor(tmp)] += 1
    
p.plot(a,x)
p.axis([0, 100, -abs(max(x)), abs(max(x))])
p.show()
```

```python
a = list(range(N))
x = list(range(N))
for i in range(N):
    x[i] = 0
for i in range(U):
    tmp1 = r.random()*N/2
    tmp2 = r.random()*N/2
    x[math.floor(tmp1 + tmp2)] += 1
    
p.plot(a,x)
p.axis([0, 100, -abs(max(x)), abs(max(x))])
p.show()
```

```python
a = list(range(N))
x = list(range(N))
for i in range(N):
    x[i] = 0
for i in range(U):
    tmp1 = r.random()*N/3
    tmp2 = r.random()*N/3
    tmp3 = r.random()*N/3
    x[math.floor(tmp1 + tmp2 + tmp3)] += 1
    
p.plot(a,x)
p.axis([0, 100, -abs(max(x)), abs(max(x))])
p.show()
```
