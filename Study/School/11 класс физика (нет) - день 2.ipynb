#### <span style="color:#55628D">Задача 4 (№ 14220)</span>
<span style="color:#55628D">По каналу связи передаются сообщения, содержащие только четыре буквы: Р, Е, К, А; для передачи используется двоичный код, удовлетворяющий условию Фано. Для букв А, Р, Е используются такие кодовые слова: А  — 111, Р  — 0, Е  — 100. Укажите кратчайшее кодовое слово для буквы К. Если таких кодов несколько, укажите код с наименьшим числовым значением. Примечание. Условие Фано означает, что ни одно кодовое слово не является началом другого кодового слова.</span>

```python
codes = ["111", "0", "100"]

def code_is_correct(c):
    for i in range(0, len(codes)):  # Пробегаемся по всем старым кодам в codes
        # Проверка того, что новый код - начало старого кода
        if codes[i][0:len(c)] == c:
            return False
            
        # Проверка того, что старый код - начало нового кода
        if codes[i] == c[0:len(codes[i])]:
            return False

    # Если не нашли "коллизий" - значит, новый код нам подходит!
    return True

# Поиск по кодам длины 1
for a1 in range(0,2):
    newcode = str(a1)
    print("Код:", newcode, "подходит:", code_is_correct(newcode))

# Поиск по кодам длины 2
for a1 in range(0,2):
    for a2 in range(0,2):
        newcode = str(a1) + str(a2)
        print("Код:", newcode, "подходит:", code_is_correct(newcode))

# Поиск по кодам длины 3
for a1 in range(0,2):
    for a2 in range(0,2):
        for a3 in range(0,2):
            newcode = str(a1) + str(a2) + str(a3)
            print("Код:", newcode, "подходит:", code_is_correct(newcode))
```

<span style="color:#55628D">1.  Строится двоичная запись числа N.</span>

<span style="color:#55628D">2.  К этой записи дописываются справа ещё два разряда по следующему правилу:</span>

<span style="color:#55628D">а)  складываются все цифры двоичной записи числа N, и остаток от деления суммы на 2 дописывается в конец числа (справа). Например, запись 11100 преобразуется в запись 111001;</span>

<span style="color:#55628D">б)  над этой записью производятся те же действия  — справа дописывается остаток от деления суммы её цифр на 2.</span>

<span style="color:#55628D">Полученная таким образом запись (в ней на два разряда больше, чем в записи исходного числа N) является двоичной записью искомого числа R. Укажите минимальное число R, которое превышает число 97 и может являться результатом работы данного алгоритма. В ответе это число запишите в десятичной системе счисления.</span>

```python
# Индексация в Python разнообразная. Она позволяет выводит элементы объекта с какого-то или по какой-то
s = "abcdef"
print(s[1:])
print(s[:4])
print(s[1:4])
```

```python
# С помощью индексации, можем убрать лишний текст с двоичного представления числа
N = 8
print(bin(N))
print(bin(N)[2:])
```

```python
# Обратно двоичное представление проводится с помощью типизации типов. Смотри на то, с какого основания образуется десятичное представление
N = 8
N2 = bin(N)[2:]
print(int(N2))
print(int(N2, 2))
print(int(N2, 3))
```

```python
# Решение задачи
def N2R(N):
    N2 = bin(N)[2:]  # Перевод в строку двоичной записи ([2:] - так мы убрали текст '0b')
    for i_1 in range(0, 2):  # Повторяем одно и то же действие дважды
        N2 += str(sum([int(i) for i in N2]) % 2)
    return(int(N2, 2))

flag = True

for i in range(20, 30):  # Проверяем среди 20 и 30 (я подобрал диапазон руками)
    r = N2R(i)  # Отображение N -> R(N)
    print("Изначальное число N =", i, "R(N) =", r, "| bin(N) = ", bin(i), "| bin(R(N)) =", bin(r))

    # Чтобы находить число "не руками"
    if flag and r >= 97:
        flag = False
        print("Правильный ответ:", r)
```

Сам по себе Python не предназначен для работы файлами Excel. Для этого нужно установить модули (библиотеки). Например, pandas (советую когда-то с ним ознакомиться).

Можете работать в самом Excel, посмотрите на YouTube как решать задачу 9 в нём.

Я скопировал 7 столбцов в файлы 1.txt ... 7.txt, и далее сделал следующее:

```python
# Файл можно открыть (utf-8) - всё копируем в блокнот
f = [open("schoolfiles/1.txt", "r"),
     open("schoolfiles/2.txt", "r"),
     open("schoolfiles/3.txt", "r"),
     open("schoolfiles/4.txt", "r"),
     open("schoolfiles/5.txt", "r"),
     open("schoolfiles/6.txt", "r"),
     open("schoolfiles/7.txt", "r")]

N = 16010
numbres = [[0, 0, 0, 0, 0, 0, 0] for i in range(N)]  # Вложенный список: 7 чисел в 16010 строке - посмотрел в таблице
for i in range(0,7):
    counter = 0
    for line in f[i]:
        a_str = line.replace("\n", "")
        if a_str != "":
            numbres[counter][i] = int(a_str)
            counter += 1

# Закрытие файлов
for i in range(0,7):
    f[i].close()

# Поиск строчек!
anw = 0  # Количество нужных строк

for i in range(0, N):
    # Отсортированный вложенный список типа: [[1, 24], [1, 32], [1, 37], [1, 49], [1, 94], [1, 115], [1, 128]]
    # Здесь, 1-е число - кол-во элементов в строке, 2-е число - элемент
    sorted_row = sorted([[numbres[i].count(a), a] for a in numbres[i]])
    
    # Проверка на то, что в строке только 2 одинаковых элемента
    if sorted_row[2][0] == 1 and sorted_row[3][0] == 2:
        # Проверка на среднее арифметическое:
        if (sorted_row[0][1] + sorted_row[1][1] + sorted_row[2][1])/3 < (sorted_row[3][1] + sorted_row[5][1])/2:
            anw += 1
            print(i, ": строка ", sorted_row)

print("Ответ:", anw)
```

#### <span style="color:#55628D">Задача 14 (№ 26959)</span>

```python
# Это решение по-нормальному
a = 16**18 * 4**10 - 4**6 - 16
a_for_compare = a  # Запись числа для проверки
print("Число в десятичной системе:", a)
s = ""

i = 0  # Номер разряда
while a > 1:
    b = a % 4  # 4 - основание системы счисления
    s = str(b) + s
    a = a // 4
    i += 1  # Увеличение номера разряда
    print("Номер разряда:", i, "| полученное число:", s)

# Проверка, что всё верно
print("Проверка:", int(s, 4) == a_for_compare)

# Ответ
print()
print("Ответ:", s.count("3"))
```
