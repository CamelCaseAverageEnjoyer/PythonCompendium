#### <span style="color:#4869bc">Иллюстрация:</span> <span style="color:#1bcaba">среднее арифметическое случайных величин имеет своё распределение</span> 

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_distribution(axs, n_possible_values: int = 100, n_events: int = int(1e6), n_to_mean: int = 1, title: str = "График") -> None:
    """Рисует график численно рассчитанного распредения среднего из N равномерно распределённых величин:
    :param axs: ось
    :param n_possible_values: Количество дискретных значений случайной величины
    :param n_events: Количество взятых событий (больше -> вернее график)
    :param n_to_mean: N величин, из которых за 1 событие берётся среднее арифметическое
    :param title: Название графика"""
    x = [0 for _ in range(n_possible_values)]
    for i in range(n_events):
        x[int(np.sum([np.random.randint(n_possible_values) for _ in range(n_to_mean)]) // n_to_mean)] += 1
    
    axs.plot([i for i in range(n_possible_values)], x)
    axs.set_ylim([0, max(x)])
    axs.set_title(title)

fig, axs = plt.subplots(1, 3, figsize=(13, 3))
for i in range(3):
    plot_distribution(axs[i], n_to_mean=i+1, title=["Равномерное распределение", "Пирамидка", "Дюна"][i])
plt.show()
```
