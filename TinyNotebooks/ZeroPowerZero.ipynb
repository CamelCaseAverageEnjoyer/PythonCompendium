#### <span style="color:#4869bc">Иллюстрация:</span> <span style="color:#1bcaba">число в степени себя $x^x$</span> 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
n = 1000
x = np.linspace(0, 0.5, n)
y = np.linspace(0, 0.5, n)
a = [[np.power(i,j) for i in x] for j in y]
a = pd.DataFrame(a, columns=x, index=y)
ax = sns.heatmap(a)
ax.invert_yaxis()
```
